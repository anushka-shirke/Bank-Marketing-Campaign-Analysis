# Project Summary: Bank Marketing Campaign Analysis

This project involves exploratory data analysis (EDA) on a bank marketing dataset to understand the factors influencing customer subscription to term deposits. The dataset includes demographic, economic, and campaign-related attributes for each client contacted during a marketing campaign.

# Key Objectives


To identify which customer attributes are most associated with a successful subscription (y = yes).

To evaluate campaign strategies, including contact methods, frequency, and duration.

To uncover trends across age, education, marital status, and more.

# Major Findings


1. Target Variable Distribution:

A significant class imbalance was observed with a majority of responses being "no" to the term deposit subscription.

2. Demographics & Subscriptions:

Job Type: Certain job roles like retired and students had higher success rates.

Education: Higher education levels showed a slightly better response rate.

Marital Status: Singles were marginally more likely to subscribe.

3. Campaign Attributes:

Contact Method: Customers contacted via cellular were more likely to subscribe compared to those contacted via telephone.

Previous Outcome: Clients with a successful outcome in a previous campaign had a significantly higher chance of subscribing again.

Month: The months of May, August, and October saw more positive responses.

4. Duration of Call:

Strongest Indicator: The longer the call duration, the higher the likelihood of a "yes". This was the most influential feature.

5. Loan, Housing, and Default Status:

Clients without loans or housing loans showed a slightly higher interest in term deposits.

6. Frequency of Contact (Campaign Count):

Lower average campaign contact count was linked to better results—too many calls reduced the success probability.

# Visual Analysis


Bar plots, scatter plots, box plots, and histograms were used to visualize trends.

Percentage-based comparisons across job, education, marital status, and more were insightful in understanding positive response likelihoods.

# Conclusion


The analysis concludes that personalized targeting based on demographics, effective contact methods (like cellular), and optimal call duration can significantly increase the chances of customer conversion in marketing campaigns. Additionally, past positive interactions are strong predictors of future success.
